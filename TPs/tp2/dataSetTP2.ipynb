```python
import wikipediaapi
import re
import csv
import pandas as pd
from google.colab import files
```

```python
# Crear el objeto Wikipedia en español
wiki = wikipediaapi.Wikipedia(user_agent='AzuTP2-AAII/1.0 (mailto:azutp2@example.com)', language='es')
```

```python
def normalize_text(text):
    # Convertir a minúsculas y quitar puntuación
    t = text.lower()
    t = re.sub(r'[\u200b-\u200f\uFEFF]', '', t)
    t = re.sub(r"[^a-záéíóúüñ0-9' ]+", ' ', t)
    t = re.sub(r'\s+', ' ', t).strip()
    return t
```

```python
def crearDataSet(nombre, temas, descargar=1):
  nombre = f"{nombre}.csv"
  with open(nombre, "w", encoding="utf-8", newline="") as f:
    writer = csv.writer(f)
    writer.writerow(["entrada", "salida"])
    count = 0

    for tema in temas:
        page = wiki.page(tema)
        if not page.exists():
            continue
        parrafos = page.text.split("\n")
        for p in parrafos:
            p = p.strip()
            if len(p) < 80:  # descartamos párrafos cortos
                continue
            entrada = normalize_text(p)
            salida = p
            writer.writerow([entrada, salida])
            count += 1
    print(f"✅ Dataset generado con {count} pares de texto.")
    df = pd.read_csv(nombre)
    if descargar:
      files.download(nombre)
    return df
```

```python
temas1 = [
    "Inteligencia artificial",
    "Argentina",
    "Matemáticas",
    "Biología",
    "Historia de la música",
    "Universidad de Buenos Aires",
    "Psicología",
    "Computadora",
    "Planeta Tierra",
    "Economía"
]
temas2 = [
    "Música",
    "Red neuronal artificial",
    "Pizza",
    "Jazmín"
]
```

Arma un df con dos columnas, la versión limpia (sin mayusculas y signos de puntuacion), y la versión original.
Cada fila es un párrafo de wikipedia en español, con título especificado en "temas".

```python
crearDataSet("train", temas1)
crearDataSet("test", temas2)
```
