```python
import wikipediaapi
import re
import csv
import pandas as pd
from google.colab import files
```

```python
# Crear el objeto Wikipedia en español
wiki = wikipediaapi.Wikipedia(user_agent='AzuTP2-AAII/1.0 (mailto:azutp2@example.com)', language='es')

# Algunos temas comunes (puedes cambiar o agregar)
temas = [
    "Inteligencia artificial",
    "Argentina",
    "Matemáticas",
    "Biología",
    "Historia de la música",
    "Universidad de Buenos Aires",
    "Psicología",
    "Computadora",
    "Planeta Tierra",
    "Economía"
]
```

```python
def normalize_text(text):
    # Convertir a minúsculas y quitar puntuación
    t = text.lower()
    t = re.sub(r'[^a-záéíóúüñ0-9() ]+', ' ', t)
    t = re.sub(r'\s+', ' ', t).strip()
    return t
```

```python
with open("dataset_wikipedia_es.csv", "w", encoding="utf-8", newline="") as f:
    writer = csv.writer(f)
    writer.writerow(["entrada", "salida"])
    count = 0

    for tema in temas:
        page = wiki.page(tema)
        if not page.exists():
            continue
        parrafos = page.text.split("\n")
        for p in parrafos:
            p = p.strip()
            if len(p) < 80:  # descartamos párrafos cortos
                continue
            entrada = normalize_text(p)
            salida = p
            writer.writerow([entrada, salida])
            count += 1
    print(f"✅ Dataset generado con {count} pares de texto.")
```
